Read Train features

```python
train_features = pd.read_csv(root_path+"train_features.csv")
train_features.head()
```

```python
train_imgs = []
size = 128
length = 487
train_sites = []

for i in range(length):
    img=cv2.imread(root_path+train_features.iloc[i, 1])
    img = cv2.resize(img, (size, size))
    img = np.array(img).flatten()
    train_imgs.append(img)
    train_sites.append(train_features.iloc[i, 2])
    
```

```python
train_imgs = np.array(train_imgs)
train_sites = np.array(train_sites)
```

```python
train_imgs.shape
```

```python
train_imgs = train_imgs/255.0
```

```python
train_imgs
```

Read train labels

```python
train_lbls = pd.read_csv(root_path+"train_labels.csv")
train_lbls = np.array(train_lbls.iloc[:487, 1:])
```

Read test features

```python
test_features = pd.read_csv(root_path+"test_features.csv")
test_features.head()
```

```python
test_features.shape
```

```python
test_imgs = []
size = 128
length = 464
test_sites = []

for i in range(length):
    img=cv2.imread(root_path+test_features.iloc[i, 1])
    img = cv2.resize(img, (size, size))
    test_imgs.append(img)
    test_sites.append(test_features.iloc[i, 2])
```

```python
test_imgs = np.array(test_imgs)
test_sites = np.array(test_sites)
```

```python
test_imgs = test_imgs/255.0
```

```python
test_imgs.shape
```

SVM

```python
train_y = []

for i in train_lbls: 
    train_y.append(list(i).index(1))

train_y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(train_imgs, train_y, test_size = 0.2, random_state = 2)

model = SVC(C=1, kernel='poly', gamma ='auto').fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred.reshape(-1, 1), y_test)
accuracy
```
